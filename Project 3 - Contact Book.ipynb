```python
# Initialize empty contact list
contacts = []

def add_contact():
  """Adds a new contact to the list"""
  name = input("Enter Name: ")
  phone = input("Enter Phone Number: ")
  email = input("Enter Email (Optional): ")
  address = input("Enter Address (Optional): ")
  contact = {"name": name, "phone": phone, "email": email, "address": address}
  contacts.append(contact)
  print("Contact added successfully!")

def view_contacts():
  """Displays all contacts in the list"""
  if not contacts:
    print("No contacts found!")
  else:
    print("Contacts:")
    for contact in contacts:
      print(f"Name: {contact['name']}, Phone: {contact['phone']}")

def search_contact():
  """Searches for a contact by name or phone number"""
  search_term = input("Enter name or phone number to search: ")
  found = False
  for contact in contacts:
    if search_term.lower() in contact["name"].lower() or search_term.lower() in contact["phone"].lower():
      print("Contact:")
      for key, value in contact.items():
        print(f"{key.title()}: {value}")
      found = True
      break
  if not found:
    print("Contact not found!")

def update_contact():
  """Updates an existing contact"""
  name = input("Enter name of contact to update: ")
  found = False
  for i, contact in enumerate(contacts):
    if contact["name"] == name:
      updated_name = input("Update name (Leave blank if no change): ") or contact["name"]
      updated_phone = input("Update phone number (Leave blank if no change): ") or contact["phone"]
      updated_email = input("Update email (Leave blank if no change): ") or contact["email"]
      updated_address = input("Update address (Leave blank if no change): ") or contact["address"]
      contacts[i] = {"name": updated_name, "phone": updated_phone, "email": updated_email, "address": updated_address}
      print("Contact updated successfully!")
      found = True
      break
  if not found:
    print("Contact not found!")

def delete_contact():
  """Deletes a contact"""
  name = input("Enter name of contact to delete: ")
  found = False
  for i, contact in enumerate(contacts):
    if contact["name"] == name:
      del contacts[i]
      print("Contact deleted successfully!")
      found = True
      break
  if not found:
    print("Contact not found!")

def main():
  """Main loop for user interaction"""
  while True:
    print("\nContact Book Menu:")
    print("1. Add Contact")
    print("2. View Contacts")
    print("3. Search Contact")
    print("4. Update Contact")
    print("5. Delete Contact")
    print("6. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
      add_contact()
    elif choice == '2':
      view_contacts()
    elif choice == '3':
      search_contact()
    elif choice == '4':
      update_contact()
    elif choice == '5':
      delete_contact()
    elif choice == '6':
      print("Exiting...")
      break
    else:
      print("Invalid choice!")

if __name__ == "__main__":
  main()
```
